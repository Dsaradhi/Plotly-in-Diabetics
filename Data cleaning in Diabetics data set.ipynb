```python
def impute_Pragnencies(cols):
    Pregnancies = cols[0]
    Outcome = cols[1]
    if pd.isnull(Pregnancies):
        if Outcome ==0:
            return 4
        elif Outcome == 1:
            return 2
        else:
            return 3
    else:
            return Pregnancies
       
```

```python
df['Pregnancies']=df[['Pregnancies','Outcome']].apply(impute_Pragnencies,axis=1)
```
